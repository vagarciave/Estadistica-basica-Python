## Actividad 3

1. Genere una matriz $A$ de ceros de dimensiones $(n\times n)$.
2. Utilizando un __While__ reemplace las columnas de la matriz $Aa$ por vectores de números aleatorios entre $-1$ y $3$.
3. Utilizando dos __While__ reemplace los elementos negativos de la matriz $A$ por $-1$ y los demás por $5$. Realice lo mismo utilizando dos bucles __for__.

```python
n = 2
A = np.zeros((n,n)) #Punto 1
aux = 0
while(aux != n): #Punto 2
    A[:,aux] = rd.uniform(-1, 3, n)
    aux+=1
print(A)
row = 0
while(row != n): #Punto 3
    col = 0
    while(col != n):
        if(A[row,col]<0):
            A[row,col] = -1
        else:
            A[row,col] = 5
        col += 1
    row += 1
print(A)
```

### Primer punto

```python
def ingresar():
    import pandas as pd
    try:
        Datos1 = pd.read_excel('Datos1.xlsx',sheat = 'Sheet1')
        existe = True
    except FileNotFoundError: #Captura la excepción, si no logra encontrar el archivo
        existe = False
    nom1 = []
    apes1 = []
    ingresa = 'SI'
    while(ingresa != 'NO'):
        nom1.append(input('Ingrese sus nombres \n').upper())
        apes1.append(input('Ingrese sus apellidos \n').upper())
        ingresa = input('¿Seguir ingresando usuario? \n').upper()
    dats = {'Nombres':nom1,'Apellidos':apes1}
    dats = pd.DataFrame(dats)
    if(not existe):
        dats.to_excel('Datos1.xlsx', index=False , header= True)
    else:
        dats = pd.concat([Datos1,dats])
        dats.to_excel('Datos1.xlsx',index=False , header= True)
```

```python
def ingresar2():
    import pandas as pd
    try:
        Datos1 = pd.read_excel('Datos2.xlsx',sheat = 'Sheet1')
        existe = True
    except FileNotFoundError: #Captura la excepción, si no logra encontrar el archivo
        existe = False
    ids = []
    nom1 = []
    nom2 = []
    apes1 = []
    apes2 = []
    edades = []
    genero = []
    gastos = []
    ingresa = 'SI'
    while(ingresa != 'NO'):
        ids.append(input('Ingrese su identificación \n').upper())
        nom1.append(input('Ingrese su primer nombre \n').upper())
        nom2.append(input('Ingrese su segundo nombre \n').upper())
        apes1.append(input('Ingrese su primer apellido \n').upper())
        apes2.append(input('Ingrese su segundo apellido \n').upper())
        edades.append(input('Ingrese su edad \n'))
        genero.append(input('Ingrese su género (Femenino o Masculino) \n').upper())
        gastos.append(input('Ingrese sus gastos aproximados por mes en pesos\n'))
        ingresa = input('¿Seguir ingresando usuario? \n').upper()
    dats = {'Id':ids,'Primer nombre':nom1,'Segundo nombre':nom2, 'Primer apellido':apes1,
           'Segundo apellido': apes2, 'Edad':edades, 'Género':genero, 'Gastos':gastos}
    dats = pd.DataFrame(dats)
    if(not existe):
        dats.to_excel('Datos2.xlsx', index=False , header= True)
    else:
        dats = pd.concat([Datos1,dats])
        dats.to_excel('Datos2.xlsx',index=False , header= True)
```

# Clase #7

En esta clase veremos como ajustar una recta para explicar la relación lineal entre dos variables cuantitativas.

```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
%matplotlib inline

#Considere que se tienen los registros de ventas en millones de euros
#durante un periodo de 9 años en una cadena de venta de ropa

#X: Inversión en publicidad(millones de euros)
#Y: Ventas(millones de euros)

x = np.array([23,26,30,34,43,48,52,57,58])
y = np.array([651,762,856,1063,1190,1298,1421,1440,1518])

plt.figure(figsize=(15,10))

plt.scatter(x,y)
plt.xlabel('x: Inversión en publicidad', fontsize=20)
plt.ylabel('y: Ventas', fontsize=20)
plt.title('Relación lineal entre x e y', fontsize=25)
plt.grid()
```

Del gráfico anterior podemos ver que a medida que aumenta la inversión en publicidad, las ventas aumentan, lo cual se puede considerar como un hecho evidente. Sin embargo, la idea es cuantificar dicha relación.

$$y_i=\beta_0+\beta_1x_i+e_i,\quad \text{para}\quad i=1, 2, \ldots, 9,$$
donde $\beta_0$ es llamado intercepto,  $\beta_1$ es llamado pendiente y $e_i$ es llamado el error.

Si despejamos los errores de la ecuación anterior, tenemos que:
$$e_i=y_i-\beta_0-\beta_1x_i$$
y la suma cuadrática de los errores se definiría como:
$$SS_e=\sum_{i=1}^9e_i^2=\sum_{i=1}^9(y_i-\beta_0-\beta_1x_i)^2$$
Se puede demostrar que los valores estimados de $\beta_0$ y $\beta_1$ que minimizan la suma $SS_e$ están dados por:
$$\left(\begin{array}{c}
\widehat{\beta}_0\\
\widehat{\beta}_1
\end{array}\right)=
\left(\left[\begin{array}{cccc}
1& 1 & \ldots & 1 \\
23& 26& \ldots & 58
\end{array}\right]
\left[\begin{array}{cc}
1& 23  \\
1& 26\\
\vdots&\vdots\\
1& 58
\end{array}\right]
\right)^{-1}
\left[\begin{array}{cccc}
1& 1 & \ldots & 1 \\
23& 26& \ldots & 58
\end{array}\right]
\left(\begin{array}{c}
651\\
762 \\
\vdots\\
1518
\end{array}\right)
$$


Para hacer estos cálculos matriciales en Phyton considere lo siguiente:

```python
ones1 = np.ones(9)
X = np.concatenate((ones1.reshape(9,1),x.reshape(9,1)), axis = 1)
Xt = X.transpose()
betas = np.linalg.inv(Xt@X)@Xt@y.reshape(9,1)
print(betas)
y_aj = betas[0]+betas[1]*x
y_aj
```

```python
plt.figure(figsize=(15,10))
plt.scatter(x,y)
plt.xlabel('x: Inversión en publicidad', fontsize=20)
plt.ylabel('y: Ventas', fontsize=20)
plt.title('Relación lineal entre x e y', fontsize=25)
plt.grid()
ecuacion = '$'+('y = %f + %f x'%(betas[0],betas[1]))+'$'; #Se genera una ecuación en LaTeX
plt.plot(x,y_aj,'-',label = ecuacion)
plt.legend(fontsize=20)
```

Los valores anteriores se interpretan como:
- Si la inversión en publicidad es igual a cero, es decir, x=0, entonces las ventas serían 167.682949 millones de euros
- Por cada unidad millón que se aumente en publicidad, es decir, por cada unidad que se aumente en x, el aumento en las ventas sería de 23.422786 millones de euros
